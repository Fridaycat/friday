```python
fig = plt.figure(figsize=(15, 7))

def plotData(plt, ax1):
  x = [p[0] for p in ax1]
  y = [p[1] for p in ax1]
  plt.plot(x, y, 'or-')

ax1 = [(2000,0.230607753), (2001,0.221382127), (2002,0.219959823), (2003,0.209839751), (2004,0.202724525),(2005,0.196328401)]

plotData(plt, ax1)


plt.show()
```
